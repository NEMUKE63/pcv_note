### numpy.arrayを使った座標の定義
im = np.array(Image.open('../data/empire.jpg'))  
とした場合、  
im[0]: 行（縦方向）の画素  
im[1]: 列（横方向）の画素  
をそれぞれ示すが、matplotlibの表示上、x軸、y軸の定義は変わらない。  
x: 横方向  
y: 縦方向  
ただし原点は左上になる。

```python
# 画像を表示する
plt.figure(figsize=(6,6))
plt.imshow(im)
# 点の座標
x = [100, 100, 400, 400]
y = [200, 500, 200, 500]
# 赤い星マークで点を描画する
plt.plot(x[0],y[0],'b*')
plt.plot(x[:2],y[:2],'b')
# タイトルを追加し、描画結果を表示する
plt.title('Plotting: "empire.jpg"')
plt.xlim(0, im.shape[1])
plt.ylim(im.shape[0],0)
plt.show()
```

```python
# 画像を表示する
plt.figure(figsize=(6,6))
plt.imshow(im)
# 点の座標
x = [100, 100, 400, 400]
y = [200, 500, 200, 500]
# 赤い星マークで点を描画する
plt.plot(x[:2],y[:2], 'r-')
# タイトルを追加し、描画結果を表示する
plt.title('Plotting: "empire.jpg"')
# 軸を消す
plt.axis('off')
plt.plot(x,y, 'r*')
plt.show()
```

### 1.2.2 画像の等高線とヒストグラム

```python
# 配列に画像を読み込む
im = np.array(Image.open('data/empire.jpg').convert('L'))
# 新しい図を作成する
plt.figure(figsize=(15,6))
plt.gray()

# オリジナルの白黒画像
plt.subplot(1,3,1)
plt.imshow(im)
plt.title('original B&W')
plt.axis('equal')
plt.axis('off')

# 左上隅を原点とする等高線を表示する
plt.subplot(1,3,2)
plt.contour(im, origin='image')
plt.title('contour')
plt.axis('equal')
plt.axis('off')

# 画像データのヒストグラム
plt.subplot(1,3,3)
plt.hist(im.flatten(),128)
plt.title('histogram')
plt.show()
```

### 1.2.3 インタラクティブな注釈

matplotlib.pyplot内の関数ginput()は画像内の指定した点の座標を抽出する。  
ただしIPython notebookではGUIバックエンドが利用できないため、Errorとなる。  
スクリプトファイルで実行すると、正しく動作する。

```python
im = np.array(Image.open('data/empire.jpg'))
plt.imshow(im)
print('３点クリックしてください')
x = plt.ginput(3)
print('クリックした座標:', x)
plt.show()
```
